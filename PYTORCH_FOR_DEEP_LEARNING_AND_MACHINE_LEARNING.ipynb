# 0. Welcome and "what is deep learning?"

Machine learning is turning things (data) into numbers and **finding patterns** in those numbers.

The computer does this part. How? Code & math. We're going to be writing the code.

Machine learning vs. Deep learning : Deep learning is a subset of Machine learning which is a subset of Artificial Intelligence.

# 1. Why use machine/deep learning?

Good reason: why not?
Better reason: for a complex problem, can you think of all the rules? ... probably not.

"If you can build a simple rule-based system that doesn't require machine learning, do that."

# 2. The number one rule of ML

What deep learning is good for :

- **Problems with long lists of rules** - when the traditional approach fails, ML/DL may help.
- **Continually changing environments** - DL can adapt ("learn") to new scenarios.
- **Discovering insights within large collections of data** - can you imagine trying to hand-craft rules for what 101 different kinds of food look like?

What deep learning is not - typically - good for :

- **When you need explainability** - the patterns learned by a DL model are typically uninterpretable by a human
- 
- **When the traditional approach is a better option** - if you can accomplish what you need with a simple rule-based system
- 
- **When errors are unacceptable** - since the outputs of DL model aren't always predictable
- 
- **When you don't have much data** - DL models usually require a fairly large amount of data to produce great results

# 3. Machine learning vs. Deep learning

- ML : gradient boosted machine algorithms, structured data
- 
- DL : neural network algorithms, unstructured data

Common algorithms : 

- ML : random forest, gradient boosted models, naive bayes, nearest neighbour, Support vector machine, ...
- 
- DL : neural networks, fully connected neural network, convolutional neural network, recurrent neural network, transformer, ...

# 4. Anatomy of neural networks

Before data gets used with a neural network, it needs to be turned into numbers.

Inputs -> Numerical encoding -> Learns representation (patterns/features/weights) -> Representation outputs -> Human-understandable outputs

Input layer (data goes in here) -> Hidden layers (learns patterns in data) -> Output layer (outputs learned representation or prediction probabilities)

Each layer is usually a combination of **linear (straight line** and/or **non-linear (not-straight line)** functions.

Note : "patterns" is an arbitrary term, you'll often hear "embedding", "weights", "feature representation", "feature vectors" all referring to similar things.

# 5. Different types of learning paradigms

- **Supervised learning** : data + labels
- 
- **Unsupervised and self-supervised learning** : data itself, without labels
- 
- **Transfer learning** : take the patterns one model learned and transfer them to another model

- (reinforcement learning)
